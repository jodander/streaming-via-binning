```python
import math

import numpy as np
import matplotlib.pyplot as plt

import efficient_bennett as eb
```

```python
# MSE approximation
max_n = 2000
factor_off = 1.02

n_vec = np.arange(30, max_n + 1, 5)
our_space = []

for n in n_vec:

    B = eb.bennett_matrix(n)
    target_mse = eb.get_mse(B, B)
    our_mse = 1000 * target_mse
    space = None

    s = 2
    while our_mse > target_mse :

        c = 1 - 1 / s

        L, R, space = eb.approx_bennett_mm_v2(n, c=c, s=s)
        our_mse = eb.get_mse(L, R)

        s += 1
        print(s, c)

    our_space.append(space)
    print(f'n={n} done w/ space={space}!')
```

```python
plt.figure()
plt.title(f"Space needed for ${factor_off}$-approximation of MSE")
plt.plot(n_vec, our_space)
plt.xlabel('n')
plt.ylabel('Space needed')
# plt.xscale('log')
plt.show()
```
